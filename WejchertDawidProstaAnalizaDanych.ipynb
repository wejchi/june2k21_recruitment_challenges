```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_excel('AirQualityUCI.xlsx',na_values=-200.)
data.isnull().sum()
```

```python
data.head()
```

```python
data.describe()
```

```python
data.columns
data = data.drop(columns=['Unnamed: 15','Unnamed: 16'])
data = data.replace(to_replace = -200, value = np.NaN)
data = data.fillna(data.median())
```

```python
data.isnull().sum()
```

```python
data.head()
```

```python
data.describe()
```

```python
data.plot(x='Date',y='RH',kind='scatter',figsize=(20,10),
        xlim=(data.iloc[0].Date + pd.DateOffset(-5),
        data.iloc[-1].Date+ pd.DateOffset(5)),
        title='wykres  wilgotności',xlabel='Data',
        ylabel='wilgotność-RH')
```

```python
daily_mean = data.groupby('Date',as_index=False).mean()
daily_mean.head()
```

```python
daily_std = data.groupby('Date',as_index=False).std()
daily_std.head()
```

```python
daily_mean.plot(x = 'Date',y = 'RH',kind = 'scatter', figsize=(15,10),
        xlim = (data.iloc[0].Date + pd.DateOffset(-5),
        data.iloc[-1].Date + pd.DateOffset(5)),
        title = 'wykres średniej dziennej wilgotności z zaznaczonym dziennym odchyleniem standardowym',
        xlabel = 'Data', ylabel = 'wilgotność-RH',
        yerr = daily_std['RH']
         )
```

```python
def outliners_v1(threshold=2.5):
    data['outlier'] = False
    for index, row in data.iterrows():
        mean = daily_mean[daily_mean['Date'] == row['Date']]['RH'].values[0]
        std = daily_std[daily_std['Date'] == row['Date']]['RH'].values[0]
        z = abs((row['RH'] - mean) / std)
        if z > threshold:
            data.loc[index,'outlier'] = True
```

```python
def outliners_v2(threshold=2.5):
    data['outlier'] = False
    for index, row in data.iterrows():
        z = abs((row['RH']-data['RH'].mean())/data['RH'].std())
        if z > threshold:
            data.loc[index,'outlier'] = True
```

```python
outliners_v1(threshold=2.)
mask = data['outlier'] == True
plt.plot(data[mask].index.values,data[mask]['RH'],'r.',label = 'outliner')
plt.plot(data[~mask].index.values,data[~mask]['RH'],'.',label = 'not outliner')
plt.legend(bbox_to_anchor=(0., 1.02, 1., .102), loc='lower left',
           ncol=2, mode="expand", borderaxespad=0.)
```

```python
outliners_v2(threshold=2.)
mask = data['outlier'] == True
plt.plot(data[mask].index.values,data[mask]['RH'],'r.',label = 'outliner')
plt.plot(data[~mask].index.values,data[~mask]['RH'],'.',label = 'not outliner')
plt.legend(bbox_to_anchor=(0., 1.02, 1., .102), loc='lower left',
           ncol=2, mode="expand", borderaxespad=0.)
```
